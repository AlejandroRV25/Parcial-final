```python
import pandas as p


#Lectura de archivo .xlsx
Arch=p.read_excel('Futbol_Partidos.xlsx')
Arch
```

```python
pais=Arch.iloc[:,7];lista_t_pais=list(pais);lista_p=set(pais);lista=list(lista_p)
repeticiones=[]
for i in lista:
    bus_pais=lista_t_pais.count(i)
    repeticiones.append(bus_pais)
df=p.DataFrame(list(zip(lista,repeticiones)),columns=['País','Partidos'])
#ista es la lista con nombres de países sin repetir.
#repeticiones es la lista con la cantidad de veces que se repite cada país.
#df es el DataFrame con lista y repeticiones organizados.
print (df)  
```

```python
import pycountry
def get_alpha_3(location):
  try:
    return pycountry.countries.get(name=location).alpha_3
  except:
    return None
df['Code'] = df['País'].apply(lambda x: get_alpha_3(x))   
print(df.head(32))
```

```python
import plotly.express as px
fig = px.choropleth(df,locations='Code',
                    color='Partidos' , hover_name='País',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
